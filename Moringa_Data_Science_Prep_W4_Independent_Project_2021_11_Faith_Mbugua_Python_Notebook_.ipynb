## Delete unwanted columns


```python
# delete the columns not neccessary for the analysis
df1=df.drop(['Displayed comment','ID','Geo point','Address','Cars'], axis=1)
df1.head()
```

## Check for number of missing entries

```python
# check the number of missing entries in the columns
df1.isnull().sum().sum()
```

## Remove duplicates

```python
# remove duplicates in the dataset
df2=df1.drop_duplicates()
df2.head()
```

## Find the difference between the blue cars at the counter

```python
df2['Diff']=df2['Bluecar counter'].diff()
df2.head()
```

## Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.


```python
# get the data for the city of Paris in April
pickup_df=df2[(df2['City']=='Paris') & (df2['month']==4)]

pickup_df.head()
```

```python
# get the 
df3=pickup_df[(pickup_df['City']=='Paris') & (pickup_df['Diff']=='')]
#df3=pickup_df[(pickup_df['Diff']=='')]

df3.sort_values(by=['Diff'],ascending=False).head()


```
